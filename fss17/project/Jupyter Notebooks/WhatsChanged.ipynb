```python
from __future__ import print_function, division

import os
import sys
from pdb import set_trace

# Update path
root = os.path.join(os.getcwd().split('project')[0], 'project')
if root not in sys.path:
    sys.path.append(root)

import numpy as np
import pandas as pd
from data.GetData import get_all_projects
from Utils.FileUtil import list2dataframe

if __name__ == "__main__":
    data = get_all_projects()
    for proj_name, proj_path in data.iteritems():
        all_changes = []
        proj_path = sorted(proj_path.data, key=lambda x:x.split('/')[-1])
        datasets = [list2dataframe(dat).dropna(axis=0).reset_index(drop=True) for dat in proj_path]
        for data_1, data_2 in zip(datasets[:-1], datasets[1:]):
            data_2 = data_2[data_2["name"].isin(data_1["name"].values.tolist())].reset_index(drop=True)
            changes = data_2[data_2.columns[1:-1]].subtract(data_1[data_1.columns[1:-1]])
            changes.set_index(changes.columns[0])
            for col in changes.columns:
                changes.loc[changes[col] == 0, col] = 0
                changes.loc[changes[col] < 0, col] = 1
                changes.loc[changes[col] > 0, col] = 1
            all_changes.append(changes.mean()*100)
        all_changes = pd.concat(all_changes, axis=1, keys=[s.name for s in all_changes])
        
        set_trace()
```
